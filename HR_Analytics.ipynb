```python
## Import All Library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
##Import Data
```

```python
hr = pd.read_csv("datasets/HR_Analytics.csv")
hr.head(10)
```

```python
##Checked Data Type & Null
```

```python
hr.info()
```

```python
##Replaced Null Value With AVG in YearWithCurrManager Column
```

```python
hr['YearsWithCurrManager'] = hr['YearsWithCurrManager'].fillna(
    hr.groupby('YearsAtCompany')['YearsWithCurrManager'].transform('mean'))
```

```python
hr.isnull().sum()
```

```python
##Changes Data Type of YearWithCurrManager from Float to Int
```

```python
hr["YearsWithCurrManager"] = hr["YearsWithCurrManager"].astype("int")
```

```python
hr.info()
```

```python
##Checked Duplicate Values 
```

```python
hr.duplicated(["EmpID"]).sum()
```

```python
##Removed Duplicates from the Data
```

```python
hr.drop_duplicates(['EmpID'],inplace=True)
hr
```

```python
hr.duplicated().sum()
```

```python
##Data Analysis
```

```python
ar = hr.groupby('Attrition').agg(Count=('Attrition', 'count'))
ar
```

## Overall Attrition

Total Employees: 1470

Attrition = Yes: 237 → ~16.1%

Attrition = No: 1233 → ~83.9%

So overall attrition in the company is about 16%.

```python
## Agegrp Wise Attrition

plt.figure(figsize=(6,4))
sns.countplot(hr, x='AgeGroup', hue="Attrition", palette='Set2')
plt.title('Attrition by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

## Attrition by Age Group

26–35 age group → Highest attrition
This group has the maximum number of people leaving, even though total employees are also high → major risk group.

18–25 age group → Moderate attrition
Young employees are leaving at a noticeable rate → likely due to career switching / better

Attrition is highest among young and mid-career employees (26–35) → company should focus retention strategies on this age group (career growth, salary, learning, promotions).

```python
# Filter only employees who left
attrition_data = hr[hr['Attrition'] == 'Yes']

# Count attrition per department
dept_counts = attrition_data['Department'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,4))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Department-wise Attrition")
plt.show()
```

## Department Wise Attrition

Research & Development – 56.1%

This department contributes the highest attrition.

👉 More than half of total employee exits are from R&D.

Insight: Possible reasons could be workload pressure, skill-market demand, role stress, or limited growth opportunities.

Business Risk: High impact area – losing technical talent affects productivity & innovation.


Sales – 38.8%

Second highest attrition contributor.

👉 Almost 2 out of every 5 employees leaving are from Sales.

Insight: Could indicate performance pressure, targets, incentives, or job stress.

Business Risk: Revenue and customer relationships may be impacted.

```python
# Filter only employees who left
attrition_data = hr[hr['Attrition'] == 'Yes']

# Plot gender-wise attrition
sns.countplot(x='Gender', data=attrition_data,palette="Set2")
plt.title("Gender-wise Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

## Gender Wise Attrition

Nearly 63% of attrition comes from male employees, which is significantly higher than females.

Male employees are almost 1.7x more likely to leave compared to female employees.

```python
plt.figure(figsize=(10,4))
plt.title('Attrition by PercentSalaryHike')
plt.xlabel('PercentSalaryHike')
plt.ylabel('Number of Employee')
sns.countplot(hr,x='PercentSalaryHike',hue='Attrition',palette="Set2")
plt.show()
```

## Attrition by PercentSalaryHike

Employees with lower hikes (11–14%) account for ~45–50% of total attrition

Employees with >20% hikes contribute <10% of attrition

Strong negative correlation between hike % and attrition

```python
plt.figure(figsize=(10,4))
plt.title('Attrition by YearsSinceLastPromotion')
plt.xlabel('YearsSinceLastPromotion')
plt.ylabel('Number of Employee')
sns.countplot(hr,x='YearsSinceLastPromotion',hue='Attrition',palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title('Attrition by SalarySlab')
plt.xlabel('SalarySlab')
plt.ylabel('Number of Employee')
sns.countplot(hr,x='SalarySlab',hue='Attrition',palette="Set2")
plt.show()
```

## Attrition by SalarySlab

The "Upto 5k" Slab
This group is the primary driver of turnover for the company.

Volume: This slab has the highest number of both total employees and leavers.

Attrition Rate: Approximately 22% (163 leavers vs. ~580 stayers).

Insight: Employees in the lowest pay bracket are significantly more likely to leave, possibly due to seeking better entry-level opportunities elsewhere or financial dissatisfaction.
